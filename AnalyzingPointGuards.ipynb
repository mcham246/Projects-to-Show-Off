**Exploring Points per game vs. Winning Percentage**

The first skill we will look at is the ability to score the basketball, this is measured by points. The graph below shows a correlation between points scored by a point guard and team wins. We are questioning the impact point guards have on today's game. With that in mind, the graph shows that 43 pecent of teams whose point guard averaged less than 20 points, won more than 50 percent of their games. Meanwhile 72 percent of teams, whose point guard averaged at least 20 points within the season won at least 50 percent of their games.

```python
#Getting the values we need, in this case points and win percentage
x = third_table['PTS'].astype(float) 
y = third_table['Win%'].astype(float)

plt.figure(figsize=(13,10))
m, b= np.polyfit(x,y, 1)

#Printing out the players on the graph
plt.plot(x, y,'.', color = 'black', label = 'Players')

#Printing out the regression line
plt.plot(x, m*x + b, color = 'red', label = 'Regression Line')
m1 = m
#Changing title, x and y axis of the table - winning
plt.title("Winning percentage vs. Points per game",fontsize=18)
plt.xlabel("Points per game",fontsize=18)
plt.ylabel("Winning percentage",fontsize=18)
plt.legend(loc="upper left")
plt.grid('on')

#Printing out the name of each player for every data on the table
for count, txt in enumerate(third_table['Player']):
        plt.annotate(txt, (x[count], y[count]), textcoords = 'offset points', xytext = (0,5), color = 'purple')
        
plt.show()
```

```python
#Getting the values we need, in this case Free throw percentage and win percentage
x = third_table['FT%'].astype(float) 
y = third_table['Win%'].astype(float)

plt.figure(figsize=(13,10))
m, b= np.polyfit(x,y, 1)
plt.plot(x, y,'.', color = 'black', label = 'Players')
plt.plot(x, m*x + b, color = 'red', label = 'Regression Line')
m8 = m

#Changing title, x and y axis of the table
plt.title("Winning percentage vs. Free throw percentage",fontsize=18)
plt.xlabel("Free throw percentage",fontsize=18)
plt.ylabel("Winning percentage",fontsize=18)
plt.legend(loc="upper left")
plt.grid('on')


#Printing out the name of each player for every data on the table
for count, txt in enumerate(third_table['Player']):
        plt.annotate(txt, (x[count], y[count]), textcoords = 'offset points', xytext = (0,5), color = 'purple')
        
plt.show()
```

**Exploring Point Guard Performance Scores to Winning Rates (2020 - 2021)**

Now that we have looked at the various skills of the point guard and the impact it has on winning. We will now give each qualifed player a total score based off their performance in points, assists, rebounds, defensive ability, EFG% and FT%. We will look at the averages for each stat for all 6 skills and see how each player performed according to the average of the rest of the players. A player with a higher score will have performed above average in more stats. We will then look at each player's given score and compare that to their team's winning percentage to see whether a point guard's performance correlates with how well a team performs or whether a point guard performance is not as valued as one might assume.

The graph below compares each point guard's recieved score with their respective team's winning percentage in the year 2020-2021. After looking at the data we can see that point guard who averaged higher scores tend to mean their team had a higher winning percentage. This is highlighted when we look at the 11 highest scored players from Trae Young to Russell Westbrook who all tend to have higher winning percentages than the rest. An outlier in this data would be Donovan Mitchell who has the highest winning percentage but stands in the very middle for score. Perhaps a coincidence, but the team he plays for, the Utah Jazz, were one of the two teams that had two point guards that started 40+ games. Perhaps meaning that two point guards that played exceptionally well is even more valued than just one, while also highlighting the value in the point guard position.

```python
#Converting to float so we can use the needed variables
third_table['PTS'] = third_table['PTS'].astype(float)
third_table['AST'] = third_table['AST'].astype(float) 
third_table['TRB'] = third_table['TRB'].astype(float) 
third_table['DFS'] = third_table['STL'].astype(float) + third_table['BLK'].astype(float)
third_table['eFG%'] = third_table['eFG%'].astype(float) 
third_table['FT%'] = third_table['FT%'].astype(float) 
third_table['score'] = 0 

#Figuring out the the score with points
for i in range(len(third_table)):
  third_table.loc[i, 'score'] = (third_table.loc[i,'PTS']/third_table['PTS'].mean())  + (third_table.loc[i,'AST']/third_table['AST'].mean()) + (third_table.loc[i,'TRB']/third_table['TRB'].mean()) + (third_table.loc[i,'DFS']/third_table['DFS'].mean()) + (third_table.loc[i,'eFG%']/third_table['eFG%'].mean()) + (third_table.loc[i,'FT%']/third_table['FT%'].mean())
```

```python
#Calculating the scores for the 2020-2021 point guards
third_table.sort_values('score', ascending=False).plot.barh(x='Player', y=['score','Win%'], color = ['red','orange'], figsize = (20,10))

#Changing some options about the graph
plt.title("Comparing 2020 - 2021 Point Guards Scores to Winning Rates",fontsize=18)
plt.xlabel("Winning Percentage",fontsize=18)
plt.ylabel("Players",fontsize=18)
```

**Exploring eFG% in 2020-2021**

The following graph specifcally focuses on what we believe is the most important trait of the point guard in today's game, efficient scoring/shooting, which is measured by their effective field goal percentage. After looking at this graph we can see that the most efficient scorers tend to lead their teams in winning percentage, as we saw from before when looking at the regression lines. This tells us that highly efficient point guards that score the basketball for their team, usually results in the team winning that game and having a successful season overall.

```python
#Multiplying the eFG% value to be able to analyze it. Numbers were too small to be analyzed.
third_table['eFG%'] = third_table['eFG%']*10
third_table.sort_values('eFG%', ascending=False).plot.barh(x='Player', y=['eFG%','Win%'], color = ['red','orange'], figsize = (20,10))

#Changing some options about the graph
plt.title("Comparing 2020 - 2021 eFG% to Winning Rates",fontsize=18)
plt.xlabel("Winning Percentage",fontsize=18)
plt.ylabel("Players",fontsize=18)
```
