```python
import folium
import requests
import pandas

arrest_table = pandas.read_csv("https://cmsc320.github.io/files/BPD_Arrests.csv")

arrest_table = arrest_table[pandas.notnull(arrest_table["Location 1"])]

arrest_table["lat"], arrest_table["long"] = arrest_table["Location 1"].str.split(",").str
arrest_table["lat"] = arrest_table["lat"].str.replace("(", "").astype(float)
arrest_table["long"] = arrest_table["long"].str.replace(")", "").astype(float)
```

```python
arrest_table.tail(20)
```

```python
import numpy 
lat = arrest_table['lat'].tolist()
long = arrest_table['long'].tolist()
incident = arrest_table['incidentOffense'].tolist()
incid_lat = {}
incid_long = {}
[]
count = 0
for items in incident:
    if items not in incid_lat:
        incid_lat[items] = [lat[count]]
    else:
        incid_lat[items].append(lat[count])
    if items not in incid_long:
        incid_long[items] = [long[count]]
    else:
        incid_long[items].append(long[count])
    count += 1
        
for items in incid_lat:
    incid_lat[items] = numpy.mean(incid_lat[items])
for items in incid_long:
    incid_long[items] = numpy.mean(incid_long[items])
```

```python

map_osm = folium.Map(location=[39.29, -76.61], zoom_start=11)
for key in incid_lat:
    folium.Marker(
        location=[incid_lat[key], incid_long[key]],
        popup=key,
        icon=folium.Icon(color="blue", icon="bullhorn"),

    ).add_to(map_osm)
map_osm
```

```python
# The map shows the average locations for each incident that occured. By clicking on a icon, it shows the charge listed. 
# I averaged the longitude and latitudes of each incident inorder to minimize the points on the map
# This way each point represents a different incident. 
# I used the averages because I can observe the general area where specific incidents occur. 
# For example buglary tends to happen more in neighborhoods while intoxicated driving incidents occur on the road.
# Because the icons all look the same the way to differentiate is by clicking on the popup which serves as a label
```
